#### After some investigation, I discovered there is a lot of duplicate data.
#### For example 9/11 is mentioned twice with same number of casualties but data is slightly different.
#### It's time to clean the data.

```python
df.duplicated().sum() # Doesn't show much because there are slight differences
```

```python
# attacks on same day at same place, by same group?
df.duplicated(['iyear', 'imonth', 'iday', 'gname', 'latitude', 'longitude']).sum()
```

```python
# it could be that same group attacked a place multiple times in a day?
# what else can we add to validate our reasoning?
# find duplicate rows based on certain columns
duplicates = df.duplicated(subset=['iyear', 'imonth', 'iday', 'gname', 'latitude', 'longitude'], keep=False)

# filter the original DataFrame using the boolean Series
duplicate_rows = df[duplicates]
```

```python
df_view = df[imp_col]
```

```python
df_view.rename(columns={'eventid': 'Events', 
                        'iyear': 'Year', 
                        'nkill': 'Killed', 
                        'nwound': 'Wounded', 
                        'country_txt': 'Country', 
                        'gname': 'Group'
                       }, inplace=True)
```

```python
df_view['Casualties'] = df_view['Killed'] + df_view['Wounded']
```

```python
df_view.iloc[73127]
```

```python
# Let's add casualties -> rationale: highly unlikely we'll have equal number of casualties in two separate attacks
# If casualty == 0; dropping such row will have no impact on our current analysis.
# If casualty is low, we can afford to drop.
# If casualty is high, definitely drop forexample two rows of 9/11 attack with 9k casualties skews the data
df_view.duplicated(['Year', 'imonth', 'iday', 'Group', 'latitude', 'longitude', 'Casualties']).sum()
```

```python
df_view.drop_duplicates(subset=['Year', 'imonth', 'iday', 'Group', 'latitude', 'longitude', 'Casualties'], keep='last', inplace=True)
```

# Visualize

```python
import altair as alt
```

```python
grouped_df = df_view.groupby(['Year', 'Group'], as_index=False)['Casualties'].sum()


# Find the top 10 terrorist groups based on total casualties
top_10_groups = grouped_df.groupby('Group')['Casualties'].sum().nlargest(10).index

# Filter the DataFrame to keep only the top 10 groups
filtered_df = grouped_df[grouped_df['Group'].isin(top_10_groups)]

# Create bar chart
alt.Chart(filtered_df).mark_bar().encode(
    x=alt.X('Year:N', title='Year'),
    y=alt.Y('sum(Casualties):Q', title='Casualties'),
    color=alt.Color('Group:N', title='Terrorist Group'),
    tooltip=['Year:N', alt.Tooltip('sum(Casualties):Q', title='Casualties')]
).properties(
    width=900,
    height=400,
    title='Casualties vs. Year for Top 10 Terrorist Groups by Number of Casualties',
    background='#F7F7F7'
).configure_axisX(
    labelAngle=90
).configure_title(
    fontSize=24,
    anchor='start', 
    color='#333333'
).configure_axis(
    labelFontSize=14,
    titleFontSize=16,
    labelColor='#333333',
    titleColor='#333333',
    grid=False
).configure_legend(
    title=None,
    orient='top-left',
    labelFontSize=14,
    titleFontSize=16,
    titleColor='#333333',
    titleLimit=0,
    labelLimit=0,
)
```
