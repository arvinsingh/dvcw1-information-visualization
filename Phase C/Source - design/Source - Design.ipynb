```python
# Numeric Py
import numpy as np
# data processing
import pandas as pd
# data visualization
import plotly.express as px
```

```python
df = pd.read_csv("globalterrorismdb_0718dist.csv", encoding="ISO-8859-1")
```

```python
# Relevant columns
df_new = df[['region_txt', 'nkill', 'nwound', 'nhostkid']]

df_new.rename(columns={'region_txt': 'Region', 
                        'nkill': 'Killed', 
                        'nwound': 'Wounded', 
                        'nhostkid': 'Kidnapped',
                       }, inplace=True)


# Group by region and calculate total for each category
df_grouped = df_new.groupby('Region').agg({
    'Killed': 'sum',
    'Wounded': 'sum',
    'Kidnapped': 'sum',
}).reset_index()

# Melt the DataFrame
df_melted = pd.melt(df_grouped, id_vars='Region', value_vars=['Killed', 'Wounded', 'Kidnapped'],
                    var_name='Incident Type', value_name='Count')

# Take the absolute value of the 'Count' column to ensure non-negative values
df_melted['Count'] = df_melted['Count'].abs()

# Create scatter plot
fig = px.scatter(
    df_melted,
    x='Region',
    y='Count',
    size='Count',
    size_max=40,
    color='Incident Type',
    labels={'Count': 'Number of Incidents'},
    title='Global Terrorism Statistics by Region (1970 - 2017)',
    color_discrete_map={'Killed': 'red', 'Wounded': 'orange', 'Kidnapped': 'blue'},  # Set custom colors
    opacity=0.75,  # Adjust marker opacity
)

# Update marker aesthetics for better visibility
fig.update_traces(marker=dict(line=dict(width=1, color='DarkSlateGray')), selector=dict(mode='markers'))


# Set axis labels and legend position
fig.update_layout(
    xaxis_title='Region',
    yaxis_title='Number of Incidents',
    font=dict(
        size=14,
    ),
    height=800,
    title=dict(font=dict(size=28)),
    yaxis=dict(
        titlefont_size=22,
        tickfont_size=16,
    ),
    xaxis=dict(
        titlefont_size=22,
        tickfont_size=16
    ),
    legend=dict(
        x=0,
        y=1,
        traceorder='normal',
        orientation='h',
    ),
)

fig.show()
```
